```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data=pd.read_csv('Clustering_gmm.csv')
```

```python
data
```

```python
db=DBSCAN(eps=1,min_samples=2)
```

```python
labels=db.fit_predict(data)
```

```python
df=pd.DataFrame(data,columns=['Weight','Height'])
df['Label']=labels
df.head()
```

```python
# print(df)
plt.figure(figsize=(8,5))
colors=['green' if label != -1 else 'red' for label in labels]
plt.scatter(data["Weight"],data["Height"], c=colors,s=100,edgecolors='black')
plt.title('DBSCAN Clustering')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.grid(True)
plt.show()
```
